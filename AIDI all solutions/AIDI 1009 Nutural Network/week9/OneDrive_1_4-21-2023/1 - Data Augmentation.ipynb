### Image Augmentation With ImageDataGenerator
We will focus on five main types of data augmentation techniques for image data; specifically:
 - Image shifts via the width_shift_range and height_shift_range arguments.
 - Image flips via the horizontal_flip and vertical_flip arguments.
 - Image rotations via the rotation_range argument
 - Image brightness via the brightness_range argument.
 - Image zoom via the zoom_range argument.
 
 <img src = "2.png">

### Horizontal and Vertical Shift Augmentation

```python
# example of horizontal shift image augmentation
from tensorflow import keras
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
%matplotlib inline  

# load the image
img = load_img('bird.jpg')
# convert to numpy array
data = img_to_array(img)
# expand dimension to one sample
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-200,200])
# prepare iterator
it = datagen.flow(samples, batch_size=1)
plt.figure(figsize=(10,4))
# generate samples and plot
for i in range(9):
    # define subplot
    plt.subplot(330 + 1 + i)
    # generate batch of images
    batch = it.next()
    # convert to unsigned integers for viewing
    image = batch[0].astype('uint8')
    # plot raw pixel data
    plt.imshow(image)
# show the figure
plt.show()
```

```python
# example of vertical shift image augmentation
datagen = ImageDataGenerator(height_shift_range=0.5)
it = datagen.flow(samples, batch_size=1)
plt.figure(figsize=(10,4))
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

### Horizontal and Vertical Flip Augmentation

```python
# example of horizontal flip image augmentation
datagen = ImageDataGenerator(horizontal_flip=True)
it = datagen.flow(samples, batch_size=1)
plt.figure(figsize=(10,4))
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

```python
# example of random rotation image augmentation
datagen = ImageDataGenerator(rotation_range=90)
it = datagen.flow(samples, batch_size=1)
plt.figure(figsize=(10,4))
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

### Random Brightness Augmentation


```python
# example of brighting image augmentation
datagen = ImageDataGenerator(brightness_range=[0.2,1.0])
it = datagen.flow(samples, batch_size=1)
plt.figure(figsize=(10,4))
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

### Random Zoom Augmentation


```python
# example of zoom image augmentation
datagen = ImageDataGenerator(zoom_range=[0.5,1.0])
it = datagen.flow(samples, batch_size=1)
plt.figure(figsize=(10,4))
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

### Tips for getting the most from image data preparation and augmentation for deep learning. 

- <strong>Review Dataset</strong>: Take note of image preparation and augmentations that might benefit the training process of your model, such as the need to handle different shifts, rotations or flips of objects in the scene.
- <strong>Review Augmentations</strong>: Review images both with individual augmentations you are using as well as the full set of augmentations you plan to use. You may see ways to simplify or further enhance your model training process.
- <strong>Evaluate a Suite of Transforms</strong>: Try more than one image data preparation and augmentation scheme. Often you can be surprised by results of a data preparation scheme you did not think would be beneficial.
