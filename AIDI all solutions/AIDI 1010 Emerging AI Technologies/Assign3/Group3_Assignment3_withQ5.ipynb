#**Question 3: Python Code (15 Marks)**

```python
import pandas as pd
import numpy as np
import seaborn as sns


from supervised.automl import AutoML

# Load the dataset
df = pd.read_csv('Nashville Properties - V2.csv')
df.tail()
```

##**Exploratory Data Analysis**

```python
# Dataset size (rows = listings, columns = features)
df.shape

# Checking for missing values
sns.heatmap(df.isna())

print(df.isnull().sum().to_string())
```

##**Data Cleaning**

```python
# Drop empty columns
df = df.dropna(axis=1, how='all')

# Fill missing values with 0 if numeric
df = df.applymap(lambda x: 0 if pd.isna(x) and isinstance(x, (int, float)) else x)
```

```python
# Handling string columns
df = df.drop(columns=[
    'listing_url',
    'name',
    'description',
    #'latitude',
    #'longitude',
    'neighborhood_overview',
    'picture_url',
    'host_id',	
    'host_url',
    'host_name',
    'host_location',
    'host_about',
    'host_response_time',
    'host_picture_url',
    'host_neighbourhood',
    'host_verifications',
    'host_has_profile_pic',
    'neighbourhood',
    'neighbourhood_cleansed',
    #'neighbourhood_group_cleansed',	
    #'license',
    'property_type', 
    'bathrooms_text',	
    'amenities', 
    'calendar_last_scraped',	
    'first_review',	
    'last_review',
    'host_since',
    #'bathrooms'
])

df['host_response_rate']    = df['host_response_rate'].str.rstrip('%').astype('float') / 100.0
df['host_acceptance_rate']  = df['host_acceptance_rate'].str.rstrip('%').astype('float') / 100.0
df['host_is_superhost'].replace({'f': 0, 't': 1}, inplace=True)
df['host_identity_verified'].replace({'f': 0, 't': 1}, inplace=True)
df['has_availability'].replace({'f': 0, 't': 1}, inplace=True)
df['instant_bookable'].replace({'f': 0, 't': 1}, inplace=True)

# Room type
df['room_type'].replace(
    {
        'Private room': 1, 
        'Entire home/apt': 2,
        'Shared room' : 3,
        'Hotel room' : 4
    }, inplace=True
)

# Price
df['price'] = df['price'].str.replace('$', '')
df['price'] = df['price'].str.replace(',', '')
df['price'] = df['price'].astype(float)
```

```python
features_to_normalize = [col for col in df.columns if col not in ['id', 'price']]

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Scale the data
scaler = MinMaxScaler()
scaler.fit(df[features_to_normalize])

# filling up missing values and creating a new Dataframe with the normalized data
df_scaled = pd.DataFrame(scaler.transform(df[features_to_normalize]), columns = features_to_normalize)
df_scaled.fillna(df_scaled.mean(), inplace=True)

df_scaled[['id', 'price']] = df[['id', 'price']]
```

##**Split Train/Test**

```python
X = df_scaled[features_to_normalize].drop(df_scaled[df_scaled['price'] > 323].index)
y = df['price'].drop(df[df['price'] > 323].index)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)

X.shape
```

##**Train and Predict**

```python
automl = AutoML(mode="Explain", train_ensemble=False)

automl.fit(X_train, y_train)
```

```python
automl.report()
```

```python
preds = automl.predict(X_test)
```

```python
preds = automl.predict(X_test)
from matplotlib import pyplot as plt

plt.plot(y_test, preds, '.')
plt.xlabel("True value")
plt.ylabel("Predicted value")
```
