# **2- ExampleA (Classification)**

```python
#Scenario: Sonar dataset, binary classification. Predict whether sonar returns indicate a rock or simulated mine.

#2.1 - Load Modules
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from hpsklearn import HyperoptEstimator
from hpsklearn import any_preprocessing #PreProcessor
from hpsklearn import any_classifier #Model
from hyperopt import tpe #Estimator
```

```python
#2.2 - Load Dataset
dfA = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/sonar.csv')

print("Data Types: \n",dfA.dtypes,"\n")
print(dfA.info(), "\n")
print("Head Records: \n",dfA.head(),"\n")
print("Shape/Data in dataframe: \n",dfA)
```

```python
#2.3.1 - Split data into input/output elements
datasetA = dfA.values
X1, y1 = datasetA[:,:-1], datasetA[:,-1]


#2.3.2 - Minimally prepare dataset
X1 = X1.astype('float32')
y1 = LabelEncoder().fit_transform(y1.astype('str')) #converting the target to string
```

```python
#2.4 - Split into train and test sets
X1_train, X1_test, y1_train, y1_test = train_test_split(X1,y1,test_size=0.33, random_state=1)
print("X1 Train: ", X1_train.shape)
print("y1 Train: ", y1_train.shape)
print("X1 Test: ", X1_test.shape)
print("y1 Test: ", y1_test.shape)
```

```python
#2.5 - Define Model/Search
hpsk_model1 = HyperoptEstimator(classifier=any_classifier('cla'),preprocessing=any_preprocessing('pre'),algo=tpe.suggest,max_evals=50,trial_timeout=30)
```

```python
#2.6 - Fit data into Model
hpsk_model1.fit(X1_train,y1_train)
```

```python
#2.7 - Summarize Performance
acc1 = hpsk_model1.score(X1_test,y1_test)
print("Accuracy: %.3f" % acc1)
```

```python
#2.8 - Summarize the best model
print(hpsk_model1.best_model())
```

# **3- ExampleB (Regression)**

```python
#Scenario: Housing dataset - predicting the house price in the suburbs of Boston

#3.1 - Load Modules
import pandas as pd
from sklearn.metrics import mean_absolute_error #MAE evaluation procedure
from sklearn.model_selection import train_test_split
from hpsklearn import HyperoptEstimator
from hpsklearn import any_preprocessing #PreProcessor
from hpsklearn import any_regressor #Model
from hyperopt import tpe #Optimizer
```

```python
#3.2 - Load Dataset
dfB = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv')

print("Data Types: \n",dfB.dtypes,"\n")
print(dfB.info(), "\n")
print("Head Records: \n",dfB.head(),"\n")
print("Shape/Data in dataframe: \n",dfB)
```

```python
#3.3 - Split data into input/output elements
datasetB = dfB.values
X2, y2 = datasetB[:,:-1], datasetB[:,-1]
print("X2 Shape",X2.shape)
print("y2 Shape",y2.shape)
```

```python
#3.4 - Split into Train & Test Sets
X2_train, X2_test, y2_train, y2_test = train_test_split(X2,y2,test_size=0.33,random_state=1)
print("X2 Train", X2_train.shape)
print("y2 Train", y2_train.shape)
print("X2 Test", X2_test.shape)
print("y2 Test", y2_test.shape)
```

```python
#3.5 - Define Model/Search
hpsk_model2 = HyperoptEstimator(regressor=any_regressor('reg'),preprocessing=any_preprocessing('pre'),loss_fn=mean_absolute_error,algo=tpe.suggest,max_evals=20,trial_timeout=30, seed=1)
```

```python
#3.6 - Fit data into Model, Perform Search
hpsk_model2.fit(X2_train,y2_train)
```

```python
#3.7 - Summarize Performance
mae2 = hpsk_model2.score(X2_test,y2_test)
print("MAE: %3.f" % mae2)
```

```python
#3.8 - Summarize the best model
print(hpsk_model2.best_model())
```
